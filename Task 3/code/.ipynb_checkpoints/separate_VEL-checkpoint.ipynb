```python
import numpy as np
import math
import matplotlib.pyplot as plt
# %matplotlib inline
import mpld3
# mpld3.enable_notebook()
```

```python
file=open("kalmann.txt",'r')
pos=np.loadtxt("kalmann.txt",delimiter=',',usecols=[0,1])
vel=np.loadtxt("kalmann.txt",skiprows=1,delimiter=',',usecols=[2,3])
vel=np.insert(vel,[0],0,axis=0)
```

```python
# # defining varibles
# del_t=1
# sigma_ax=0.1
# sigma_ay=0.1
# sigma_x=0.3
# sigma_y=0.3
# sigma_vx=10
# sigma_vy=10
# pos_x0,pos_y0=pos[0]
# p=10
```

```python
# defining varibles
del_t=1
sigma_ax=1
sigma_ay=1
sigma_x=0.1
sigma_y=0.1
sigma_vx=0.01
sigma_vy=0.01
pos_x0,pos_y0=pos[0]
p=10
```

```python
pred_statex=[]
est_statex=[]
pred_px=[]
est_px=[]
```

```python
pred_statey=[]
est_statey=[]
pred_py=[]
est_py=[]
```

```python
F=np.zeros((3,3))
F[0,:3]=(1,del_t,0.5*del_t**2)
F[1,:3]=(0,1,del_t)
F[2,2]=1
# F[3,3:6]=(1,del_t,0.5*del_t**2)
# F[4,3:6]=(0,1,del_t)
# F[5,5]=1
FT=np.transpose(F)
# F
```

```python
# Q=np.zeros((6,6))
# Q[0,:3]=(0.05*del_t**5*sigma_ax**2,0.125*del_t**4*sigma_ax**2,(1/6)*del_t**3*sigma_ax**2)
# Q[1,:3]=(0,(1/3)*del_t**3*sigma_ax**2,0.5*del_t**2*sigma_ax**2)
# Q[2,:3]=(0,0,1*sigma_ax**2*del_t)
# Q[3,3:6]=(0.05*del_t**5*sigma_ay**2,0.125*del_t**4*sigma_ay**2,(1/6)*del_t**3*sigma_ay**2)
# Q[4,3:6]=(0,(1/3)*del_t**3*sigma_ay**2,0.5*del_t**2*sigma_ay**2)
# Q[5,3:6]=(0,0,1*sigma_ay**2*del_t)
# # Q=Q*sigma_ay**2
# # Q-Continuos noise
```

```python
Qx=np.zeros((3,3))
Qx[0,:3]=(0.25*del_t**4*sigma_ax**2,0.5*del_t**3*sigma_ax**2,0.5*del_t**2*sigma_ax**2)
Qx[1,:3]=(0,del_t**2*sigma_ax**2,del_t*sigma_ax**2)
Qx[2,:3]=(0,0,1*sigma_ax**2)
# Q[3,3:6]=(0.25*del_t**4*sigma_ay**2,0.5*del_t**3*sigma_ay**2,0.5*del_t**2*sigma_ay**2)
# Q[4,3:6]=(0,del_t**2*sigma_ay**2,del_t*sigma_ay**2)
# Q[5,3:6]=(0,0,1*sigma_ay**2)
# # Q=Q*sigma_ay**2
# Q-Discrete noise
```

```python
Qy=np.zeros((3,3))
Qy[0,:3]=(0.25*del_t**4*sigma_ay**2,0.5*del_t**3*sigma_ay**2,0.5*del_t**2*sigma_ay**2)
Qy[1,:3]=(0,del_t**2*sigma_ay**2,del_t*sigma_ay**2)
Qy[2,:3]=(0,0,1*sigma_ay**2)
# Q[3,3:6]=(0.25*del_t**4*sigma_ay**2,0.5*del_t**3*sigma_ay**2,0.5*del_t**2*sigma_ay**2)
# Q[4,3:6]=(0,del_t**2*sigma_ay**2,del_t*sigma_ay**2)
# Q[5,3:6]=(0,0,1*sigma_ay**2)
# # Q=Q*sigma_ay**2
# Q-Discrete noise
```

```python
Rx=np.zeros((2,2))
Rx[0,0],Rx[1,1]=(sigma_x**2,sigma_vx**2)
# R
```

```python
Ry=np.zeros((2,2))
Ry[0,0],Ry[1,1]=(sigma_y**2,sigma_vy**2)
# R
```

```python
H=np.zeros((2,3))
H[0,0],H[1,1] =(1,1)
HT=np.transpose(H)
# H
```

```python
temp=np.zeros((3,1))
temp[:,0]=[pos_x0,0,0]
est_statex.append(temp)
# est_state[0]
```

```python
temp=np.zeros((3,1))
temp[:,0]=[pos_y0,0,0]
est_statey.append(temp)
# est_state[0]
```

```python
temp=np.diag((p,p,p))
est_px.append(temp)
# est_p[0]
```

```python
temp=np.diag((p,p,p))
est_py.append(temp)
# est_p[0]
```

```python
def state_extrapolate(xn):
    xnext=np.dot(F,xn)
    return xnext
    
```

```python
def cov_extrapolate(pn):
    pnext=np.dot(np.dot(F,pn),FT) 
    return pnext
```

```python
def measurex(n):
    z=np.zeros((2,1))
    z[:,0]=(pos[n,0],vel[n,0])
    return z
    
```

```python
def measurey(n):
    z=np.zeros((2,1))
    z[:,0]=(pos[n,1],vel[n,1])
    return z
    
```

```python
def K_gainx(n):
    kn=np.dot(np.dot(pred_px[n-1],HT),(np.linalg.inv(np.dot(np.dot(H,pred_px[n-1]),HT)+Rx)))
    return kn
              
```

```python
def K_gainy(n):
    kn=np.dot(np.dot(pred_py[n-1],HT),(np.linalg.inv(np.dot(np.dot(H,pred_py[n-1]),HT)+Ry)))
    return kn
              
```

```python
def state_update(xn_1,kn,zn):
    xn=xn_1 + np.dot(kn,(zn-np.dot(H,xn_1)))
    return xn
```

```python
def cov_updatex(pn_1,kn):
    idm=(np.identity(3)-np.dot(kn,H))
    idmT=np.transpose(idm)
    pn=np.dot(np.dot(idm,pn_1),idmT)+ np.dot(np.dot(kn,Rx),np.transpose(kn))
    return pn
```

```python
def cov_updatey(pn_1,kn):
    idm=(np.identity(3)-np.dot(kn,H))
    idmT=np.transpose(idm)
    pn=np.dot(np.dot(idm,pn_1),idmT)+ np.dot(np.dot(kn,Ry),np.transpose(kn))
    return pn
```

```python
pred_statex.append(state_extrapolate(est_statex[0]))
pred_statey.append(state_extrapolate(est_statey[0]))
pred_px.append(cov_extrapolate(est_px[0])+Qx)
pred_py.append(cov_extrapolate(est_py[0])+Qy)
for i in range(1,len(vel)):
    print("\n---------------------------------\n\nIteration :-",i)
    zix=measurex(i)
    ziy=measurey(i)
    kix=K_gainx(i)
    kiy=K_gainy(i)
    est_statex.append(state_update(pred_statex[i-1],kix,zix))
    est_statey.append(state_update(pred_statey[i-1],kiy,ziy))
    print("Estimated position(x,y) :- (%g,%g)" % (est_statex[i][0],est_statey[i][0]))
    est_px.append(cov_updatex(pred_px[i-1],kix))
    est_py.append(cov_updatey(pred_py[i-1],kiy))
    pred_statex.append(state_extrapolate(est_statex[i]))
    pred_statey.append(state_extrapolate(est_statey[i]))
    print("Uncertainity(x,y) :- (%g,%g)" % (est_px[i][0,0],est_py[i][0,0]))
    pred_px.append(cov_extrapolate(est_px[i]))
    pred_py.append(cov_extrapolate(est_py[i]))
    
    
```

```python
i=np.arange(1,len(vel))
ar=[]
br=[]
m=[]
for p in range(1,len(vel)):
    ar.append(pred_statex[p][0])
    br.append(est_statex[p][0])
    m.append(pos[p][0])
# pred_state[2][0]
# plt.plot(i,ar,color='orange')
plt.plot(i,br,color='blue')
plt.plot(i,m,color='red')
plt.title("comparision of estimated x(blue) and measured x(red)")
plt.show()
# comparision of estimated x(blue) and measured x(red)
```
