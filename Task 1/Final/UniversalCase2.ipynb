```python
import math
import time
from PIL import Image
import numpy as np
im = Image.open("Task_1_Low.png")
np_im = np.array(im)
width,height = im.size
```

```python
def find_start():
    for x in range(height):
        for y in range(width):
            a,b,c=np_im[x,y]
            if((a,b,c)==(45,204,113)):
                return x,y
```

```python
def find_end():
    for x in range(height):
        for y in range(width):
            a,b,c=np_im[x,y]
            if((a,b,c)==(231,76,60)):
                return x,y
```

```python
def min_index(list,arr):
    min = arr[list[0]]
    mini = 0
    for i in range(len(list)):
        if arr[list[i]]<=min:
            min=arr[list[i]]
            mini=i
    return mini
```

```python
def h(mode,x,y):
    if(mode=='m'):
        return abs(p-x) + abs(q-y)
    if(mode=='e'):
        return math.sqrt(abs(p-x)**2 + abs(q-y)**2)
    if(mode=='dg'):
        return abs(p-x) + abs(q-y) -1*(2-1.4)* min(abs(p-x),abs(q-y))
    if(mode=='dik'):
        return 0
    if(mode=='adm'):
        return min(abs(p-x) ,abs(q-y))
    if(mode=='nonadm'):
        return 1000*(abs(p-x) + abs(q-y))
#         h=min(abs(p-x) ,abs(q-y))+0.01
#         return (0.01*math.log(h) +1.2)*h
```

```python
def path(a,b):
    while 1:
        x,y,z = np_im[a,b]
        if (x,y,z) ==(45,204,113):
            break
        np_im[a,b]=(0,255,0)
        if(parent[a,b]==1):
            a=a-1
        elif (parent[a,b]==2):
            b=b+1
        elif (parent[a,b]==3):
            a=a+1
        elif (parent[a,b]==4):
            b=b-1
        elif (parent[a,b]==5):
            a=a-1
            b=b+1
        elif (parent[a,b]==6):
            a=a+1
            b=b+1
        elif (parent[a,b]==7):
            a=a+1
            b=b-1
        elif (parent[a,b]==8):
            a=a-1
            b=b-1
        
```

```python
def resize(image):
    new_image=[]
    for i in range(100):
        for l in range(10):
            for j in range(100):
                a,b,c=image[i,j]          
                for k in range(10):
                    new_image.append((a,b,c))
    new_image=np.array(new_image)
    new_image=new_image.reshape((1000,1000,3))
    new_image=Image.fromarray(new_image)
#     new_image.save("resize.png")    
    return new_image
```

```python
x,y=find_start()
p,q=find_end()
parent = np.zeros((height,width))
dist = np.ones((height,width))*10000000
f = np.ones((height,width))*10000000
dist[x,y]=0
f[x,y]=0
open = [(x,y)]
```

```python
# Enter Mode of Operation
# ------------------------------
# Mode            Keyword

# Manhattan         m
# Euclidian         e
# Diagonal          dg
# Dijsktra          dik
# Admissable        adm
# Non-Admisable     nonadm
mode='nonadm'
```

```python
start= time.time()
fgh(mode)
path(p,q)
end=time.time()
s= end-start
print("Time=",s)
print("Cost=",dist[p,q])
```
