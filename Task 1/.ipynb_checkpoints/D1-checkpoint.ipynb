```python
from PIL import Image
import numpy as np
im = Image.open("Task_1_Low.png")
np_im = np.array(im)
width,height = im.size
```

```python
def find_start():
    for x in range(width):
        for y in range(height):
            a,b,c=np_im[x,y]
            if((a,b,c)==(45,204,113)):
                return x,y
```

```python
def find_end():
    for x in range(width):
        for y in range(height):
            a,b,c=np_im[x,y]
            if((a,b,c)==(231,76,60)):
                return x,y
```

```python
def min_index(list):
    min = dist[list[0]]
    mini = 100000
    for i in range(len(list)):
        if dist[list[i]]<=min:
            min=dist[list[i]]
            mini=i
    return mini
        
```

```python
x,y=find_start()
```

```python
trvld = np.zeros((width,height))
dist = np.ones((width,height))*10000000
dist[x,y]=0
open = [(x,y)]
```

```python
while len(open):
    print(open)
    c_x,c_y = open[min_index(open)]
    open.pop(min_index(open))
    if (c_x+1)<width:
        a,b,c=np_im[c_x+1,c_y]
        if (dist[c_x+1,c_y]>dist[c_x,c_y]+1) and ((a,b,c)!=(255,255,255)):
            dist[c_x+1,c_y]=dist[c_x,c_y]+1
            open.append((c_x+1,c_y))
            np_im[c_x+1,c_y]=(255,0,0)
        if (a,b,c)==[231,76,60]:
            break
    if (c_x-1)>=0:
        a,b,c=np_im[c_x-1,c_y]
        if (dist[c_x-1,c_y]>dist[c_x,c_y]+1) and ((a,b,c)!=(255,255,255)):
            dist[c_x-1,c_y]=dist[c_x,c_y]+1
            open.append((c_x-1,c_y))
            np_im[c_x-1,c_y]=(255,0,0)
        if (a,b,c)==[231,76,60]:
            break
    if (c_y+1)<width:
        a,b,c=np_im[c_x,c_y+1]
        if (dist[c_x,c_y+1]>dist[c_x,c_y]+1) and ((a,b,c)!=(255,255,255)):
            dist[c_x,c_y+1]=dist[c_x,c_y]+1
            open.append((c_x,c_y+1))
            np_im[c_x,c_y+1]=(255,0,0)
        if (a,b,c)==[231,76,60]:
            break
    if (c_y-1)>=0:
        a,b,c=np_im[c_x,c_y-1]
        if (dist[c_x,c_y-1]>dist[c_x,c_y]+1) and ((a,b,c)!=(255,255,255)):
            dist[c_x,c_y-1]=dist[c_x,c_y]+1
            open.append((c_x,c_y-1))
            np_im[c_x,c_y-1]=(255,0,0)
        if (a,b,c)==[231,76,60]:
            break
#     new_im = Image.fromarray(np_im)
#     new_im.show()
    
```
